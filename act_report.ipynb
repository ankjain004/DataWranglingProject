```python
import matplotlib.pyplot as plt
import pandas as pd
twitter_archive_master = pd.read_csv('twitter_archive_master.csv')
pd.value_counts(twitter_archive_master['Dog_Stage']).plot.bar()
plt.title("Dog Stage & Count")
plt.xlabel("Dog Stage Names")
plt.ylabel("Count on each Dog Stage")
```

In above graph, our objective is showcase each dog stage with their respective count.
In X-axix, we used column "Dog_Stage" and in Y axix, we see their count value.
We can easily say that Puuper is famous dog stage and approx 70-75% Dog stage is NONE.
